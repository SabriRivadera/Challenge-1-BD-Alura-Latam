### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 2. Ventas por categoría

##Ingreso total por cada tienda

```python
# Calculamos los ingresos totales de cada tienda sumando la columna precio.

ingreso_total_por_tienda = tienda["Precio"].sum().astype(int)
ingreso_total_por_tienda_2 = tienda2["Precio"].sum().astype(int)
ingreso_total_por_tienda_3 = tienda3["Precio"].sum().astype(int)
ingreso_total_por_tienda_4 = tienda4["Precio"].sum().astype(int)
print("el ingreso total de la tienda 1 es:", ingreso_total_por_tienda)
print("el ingreso total de la tienda 2 es:", ingreso_total_por_tienda_2)
print("el ingreso total de la tienda 3 es:", ingreso_total_por_tienda_3)
print("el ingreso total de la tienda 4 es:", ingreso_total_por_tienda_4)
```

```python
iendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
calificaciones = [
    ingreso_total_por_tienda,
    ingreso_total_por_tienda_2,
    ingreso_total_por_tienda_3,
    ingreso_total_por_tienda_4
]

colores = ["#355C7D", "#5B9A95", "#7FB099", "#B7C9A8"]

plt.pie(
    calificaciones,
    labels=tiendas,
    autopct="%1.1f%%",
    startangle=90,
    colors=colores
)

plt.title("Proporción de ingreso por tienda")

plt.show()
```

```python
# Calculamos la cantidad de productos vendidos de cada categoria en cada tienda
ventas_categoria_tienda = tienda.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
ventas_categoria_tienda_2 = tienda2.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
ventas_categoria_tienda_3 = tienda3.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
ventas_categoria_tienda_4 = tienda4.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
```

# 3. Calificación promedio de la tienda


```python
#Calculamos la satisfaccion promedio del cliente con los productos vendidos

calificacion_tienda = tienda["Calificación"].mean()
calificacion_tienda_2 = tienda2["Calificación"].mean()
calificacion_tienda_3 = tienda3["Calificación"].mean()
calificacion_tienda_4 = tienda4["Calificación"].mean()

print("la satisfaccion promedio del cliente con los productos vendidos de la tienda 1 es:", calificacion_tienda)
print("la satisfaccion promedio del cliente con los productos vendidos de la tienda 2 es:", calificacion_tienda_2)
print("la satisfaccion promedio del cliente con los productos vendidos de la tienda 3 es:", calificacion_tienda_3)
print("la satisfaccion promedio del cliente con los productos vendidos de la tienda 4 es:", calificacion_tienda_4)
```

```python
import seaborn as sns

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
calificaciones = [
    calificacion_tienda,
    calificacion_tienda_2,
    calificacion_tienda_3,
    calificacion_tienda_4
]

colores = sns.color_palette("Blues", 4)

plt.bar(tiendas, calificaciones, color=colores)

plt.title("Calificación promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación promedio")

plt.show()
```

# 4. Productos más y menos vendidos

```python
productos_mas_tienda = tienda["Producto"].value_counts().head()
productos_mas_tienda_2 = tienda2["Producto"].value_counts().head()
productos_mas_tienda_3 = tienda3["Producto"].value_counts().head()
productos_mas_tienda_4 = tienda4["Producto"].value_counts().head()

fig, axes = plt.subplots(4, 1, figsize=(8, 16))

color_azul = "#3C7A89"

productos_mas_tienda.plot(color=color_azul, kind="barh", ax=axes[0], title="Top 5 productos más vendidos - Tienda 1")
productos_mas_tienda_2.plot(color=color_azul, kind="barh", ax=axes[1], title="Top 5 productos más vendidos - Tienda 2")
productos_mas_tienda_3.plot(color=color_azul, kind="barh", ax=axes[2], title="Top 5 productos más vendidos - Tienda 3")
productos_mas_tienda_4.plot(color=color_azul, kind="barh", ax=axes[3], title="Top 5 productos más vendidos - Tienda 4")

for ax in axes:
    ax.invert_yaxis()
    ax.set_xlabel("Cantidad vendida")
    ax.set_ylabel("Producto")

plt.tight_layout()
plt.show()
```

```python
productos_menos_tienda = tienda["Producto"].value_counts().tail()
productos_menos_tienda_2 = tienda2["Producto"].value_counts().tail()
productos_menos_tienda_3 = tienda3["Producto"].value_counts().tail()
productos_menos_tienda_4 = tienda4["Producto"].value_counts().tail()

fig, axes = plt.subplots(4, 1, figsize=(8, 16))

color_rojo = "#B56576"

productos_menos_tienda.plot(color=color_rojo, kind="barh", ax=axes[0], title="Top 5 productos menos vendidos - Tienda 1")
productos_menos_tienda_2.plot(color=color_rojo, kind="barh", ax=axes[1], title="Top 5 productos menos vendidos - Tienda 2")
productos_menos_tienda_3.plot(color=color_rojo, kind="barh", ax=axes[2], title="Top 5 productos menos vendidos - Tienda 3")
productos_menos_tienda_4.plot(color=color_rojo, kind="barh", ax=axes[3], title="Top 5 productos menos vendidos - Tienda 4")

for ax in axes:
    ax.invert_yaxis()
    ax.set_xlabel("Cantidad vendida")
    ax.set_ylabel("Producto")

plt.tight_layout()
plt.show()
```
